## Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.

Ans= The main difference between a t-test and a z-test lies in the assumptions about the population standard deviation. Here's a breakdown of the differences and an example scenario for each type of test:

t-Test:

1) Assumption: The t-test assumes that the population standard deviation is unknown and must be estimated from the sample.
2) Sample Size: The t-test is appropriate for small sample sizes (typically less than 30) or when the population standard deviation is unknown.
3) Distribution: The t-test uses the t-distribution to calculate critical values and p-values.
4) Variance: The t-test accounts for the additional uncertainty introduced by estimating the population standard deviation based on the sample.

Example Scenario: Suppose you want to compare the average test scores of two groups of students. You have two independent samples, each with less than 30 students, and you do not know the population standard deviation. In this case, you would use a t-test to assess whether the means of the two groups are statistically different.

z-Test:

1) Assumption: The z-test assumes that the population standard deviation is known or that the sample size is large enough for the sample standard deviation to be a good estimate of the population standard deviation.
2) Sample Size: The z-test is appropriate for large sample sizes (typically greater than 30) or when the population standard deviation is known.
3) Distribution: The z-test uses the standard normal distribution (z-distribution) to calculate critical values and p-values.
4) Variance: The z-test does not account for the additional uncertainty of estimating the population standard deviation since it assumes it is known or approximated accurately.

Example Scenario: Suppose you want to assess whether a new drug has an impact on blood pressure. You conduct a clinical trial with a large sample of 500 participants, and the population standard deviation for blood pressure is known. In this case, you would use a z-test to compare the mean blood pressure before and after administering the drug.

## Q2: Differentiate between one-tailed and two-tailed tests.

Ans= One-Tailed Test:

A one-tailed test, also known as a one-sided test, is a statistical test where the alternative hypothesis is specified in one direction. It focuses on detecting an effect or difference in a specific direction. In a one-tailed test, the critical region is located entirely in one tail of the distribution.

Example:
Suppose a company claims that their new product increases customer satisfaction. A one-tailed test would focus on testing if the new product improves satisfaction (alternative hypothesis: mean satisfaction is greater than a specified value), without considering the possibility of decreased satisfaction.

Hypotheses:

Null Hypothesis (H0): The mean satisfaction is equal to or less than a specified value.

Alternative Hypothesis (Ha): The mean satisfaction is greater than a specified value.

Critical Region: The critical region for a one-tailed test would be located entirely in the upper tail of the distribution.

Two-Tailed Test:

A two-tailed test, also known as a two-sided test, is a statistical test where the alternative hypothesis is specified in both directions. It aims to detect any difference or effect, regardless of the direction. In a two-tailed test, the critical region is divided equally between the two tails of the distribution.

Example:
Suppose a researcher wants to determine if there is a significant difference in test scores between two groups. A two-tailed test would consider the possibility of a difference in either direction, without assuming which group has higher scores.

Hypotheses:

Null Hypothesis (H0): The means of the two groups are equal.

Alternative Hypothesis (Ha): The means of the two groups are different.

Critical Region:The critical region for a two-tailed test would be divided between the upper and lower tails of the distribution, reflecting the possibility of differences in either direction.


## Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.

Ans= Type I Error:
A Type I error occurs when the null hypothesis is incorrectly rejected, even though it is true. It is also known as a false positive. This means that the test incorrectly indicates a significant effect or difference when there is none in reality.

Example Scenario:
Suppose a medical researcher is conducting a clinical trial to test the effectiveness of a new drug. The null hypothesis (H0) states that the drug has no effect on the condition being treated. However, due to random chance or other factors, the researcher observes a significant effect in the data and rejects the null hypothesis. In reality, the drug has no actual effect, and the observed significant effect is a Type I error.

Type II Error:
A Type II error occurs when the null hypothesis is incorrectly failed to be rejected, even though it is false. It is also known as a false negative. This means that the test fails to detect a significant effect or difference when one exists.

Example Scenario:
Continuing with the previous example, suppose the new drug actually has a significant effect on the condition being treated, but the clinical trial fails to detect it. The null hypothesis (H0) is not rejected, leading to a Type II error. In this case, patients may miss out on a potentially effective treatment due to the failure to identify the true effect of the drug.

## Q4: Explain Bayes's theorem with an example.

Ans= Bayes's theorem is a fundamental concept in probability theory that describes how to update the probability of a hypothesis or event based on new evidence. It allows us to revise our beliefs or estimates as new information becomes available. The theorem is named after Thomas Bayes, an 18th-century mathematician.

Bayes's theorem states:

P(A|B) = (P(B|A) * P(A)) / P(B)

Where:

P(A|B) represents the probability of event A occurring given that event B has occurred.

P(B|A) represents the probability of event B occurring given that event A has occurred.

P(A) and P(B) represent the probabilities of events A and B occurring independently.

Example: Suppose you have two boxes, Box A and Box B. Box A contains 4 red balls and 1 green ball, while Box B contains 3 red balls and 2 green balls. You randomly select one box and then randomly draw a ball from the chosen box. You draw a red ball.

Now, let's assign the events:

A: You selected Box A.

B: You drew a red ball.

We want to find the probability that you selected Box A given that you drew a red ball.

We are given:

P(A) = 0.5 (since you randomly selected one box out of two)

P(B|A) = 4/5 (probability of drawing a red ball given that you selected Box A)

Using Bayes's theorem:

P(A|B) = (P(B|A) * P(A)) / P(B)

P(A|B) = (4/5 * 0.5) / P(B)

To calculate P(B), we need to consider both cases: drawing a red ball from Box A and drawing a red ball from Box B.

P(B) = P(B|A) * P(A) + P(B|~A) * P(~A)

P(B) = (4/5 * 0.5) + (3/5 * 0.5)

Substituting the values into Bayes's theorem:

P(A|B) = (4/5 * 0.5) / [(4/5 * 0.5) + (3/5 * 0.5)]

Calculating this expression will give us the revised probability of having selected Box A given that you drew a red ball.

## Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

Ans= A confidence interval is a range of values that is used to estimate an unknown population parameter, such as the mean or proportion. It provides a measure of uncertainty around the point estimate and gives us a sense of how precise our estimate is.

An example to illustrate how to calculate a confidence interval:

Suppose we want to estimate the average height of a certain population. We collect a random sample of 100 individuals from that population and measure their heights. The sample mean height is 170 cm, and the sample standard deviation is 5 cm.

Point Estimate:
The point estimate for the population mean height is the sample mean, which is 170 cm.

Standard Error:
The standard error is calculated as the sample standard deviation divided by the square root of the sample size:

Standard Error = sample standard deviation / √sample size

Standard Error = 5 cm / √100

Standard Error = 5 cm / 10

Standard Error = 0.5 cm

Critical Value:
The critical value depends on the desired confidence level and the distribution associated with the sample statistic. Let's assume we want a 95% confidence interval, which corresponds to a critical value of approximately 1.96 for a large sample size (using the standard normal distribution).

Calculate the Confidence Interval:
The confidence interval is calculated by multiplying the standard error by the critical value and then adding/subtracting the result from the point estimate:

Confidence Interval = Point Estimate ± (Critical Value * Standard Error)

Confidence Interval = 170 cm ± (1.96 * 0.5 cm)

The calculated confidence interval would be:
Confidence Interval = (169.02 cm, 170.98 cm)

Interpretation:
We can be 95% confident that the true population mean height lies within the range of 169.02 cm to 170.98 cm based on the sample data.

## Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the event's probability and new evidence. Provide a sample problem and solution.

Ans= Suppose a factory produces two types of products: Product A and Product B. It is known that 60% of the products are of type A, while the remaining 40% are of type B. Product A has a defect rate of 5%, while Product B has a defect rate of 10%. If a randomly selected product is found to be defective, what is the probability that it is of type B?

Solution:
Let's assign the events as follows:

A: The product is of type A.

B: The product is of type B.

D: The product is defective.

We want to find P(B|D), the probability that the product is of type B given that it is defective.

We are given:

P(A) = 0.60 (probability of selecting a product of type A)

P(B) = 0.40 (probability of selecting a product of type B)

P(D|A) = 0.05 (probability of a defective product given it is of type A)

P(D|B) = 0.10 (probability of a defective product given it is of type B)

We can use Bayes' Theorem to calculate P(B|D):

P(B|D) = (P(D|B) * P(B)) / P(D)

To calculate P(D), we need to consider both types of products:

P(D) = P(D|A) * P(A) + P(D|B) * P(B)

Substituting the given probabilities into Bayes' Theorem:

P(B|D) = (P(D|B) * P(B)) / (P(D|A) * P(A) + P(D|B) * P(B))

P(B|D) = (0.10 * 0.40) / ((0.05 * 0.60) + (0.10 * 0.40))

Simplifying the expression:

P(B|D) = 0.04 / (0.03 + 0.04)

P(B|D) = 0.04 / 0.07

P(B|D) ≈ 0.571

Therefore, the probability that a defective product is of type B is approximately 0.571 or 57.1%.

## Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is significantly effective at a 95% confidence level using a t-test.

Ans= Step 1: State the hypotheses:

Null Hypothesis (H0): The average weight loss is not significantly different from zero (no effect of the drug): μ = 0.

Alternative Hypothesis (Ha): The average weight loss is significantly different from zero (effect of the drug): μ ≠ 0.

Step 2: Set the significance level:
The significance level (α) is given as 0.05, which corresponds to a 95% confidence level.

Step 3: Calculate the test statistic:
Since the population standard deviation is unknown, we will use a t-test. The test statistic for a one-sample t-test is calculated as:

t = (x̄ - μ) / (s / √n)

Where:

In this example:

x̄ = 6 (sample mean)

μ = 0 (hypothesized population mean)

s = 2.5 (sample standard deviation)

n = 50 (sample size)

Calculating the test statistic:

t = (6 - 0) / (2.5 / √50)

Step 4: Determine the critical value:
Since this is a two-tailed test (we are testing for a significant difference in either direction), we need to find the critical values for a two-tailed t-test at the specified significance level (α = 0.05) and degrees of freedom (n - 1 = 49). Using a t-table or statistical software, the critical values are approximately ±2.009.

Step 5: Make a decision:
Compare the absolute value of the test statistic to the critical value. If the test statistic falls outside the critical region (i.e., if |t| > critical value), we reject the null hypothesis. Otherwise, we fail to reject the null hypothesis.

## Q12. A researcher is testing the effectiveness of two different teaching methods on student performance. Sample A has a mean score of 85 with a standard deviation of 6, while sample B has a mean score of 82 with a standard deviation of 5. Conduct a hypothesis test to determine if the two teaching methods have a significant difference in student performance using a t-test with a significance level of 0.01.

Ans= Step 1: State the hypotheses:

Null Hypothesis (H0): The two teaching methods have no significant difference in student performance. μA = μB.

Alternative Hypothesis (Ha): The two teaching methods have a significant difference in student performance. μA ≠ μB.

The significance level (α) is given as 0.01, which corresponds to a 99% confidence level.

Step 3: Calculate the test statistic:
Since we are comparing the means of two independent samples, we can use a two-sample t-test. The test statistic for a two-sample t-test is calculated as:

t = (x̄A - x̄B) / √((sA^2 / nA) + (sB^2 / nB))

Where:

x̄A = 85 (sample mean for sample A)
x̄B = 82 (sample mean for sample B)
sA = 6 (sample standard deviation for sample A)
sB = 5 (sample standard deviation for sample B)
nA = nB = (sample sizes for both samples)

Calculating the test statistic:

t = (85 - 82) / √((6^2 / nA) + (5^2 / nB))

Step 4: Determine the critical value:
Since this is a two-tailed test (we are testing for a significant difference in either direction), we need to find the critical value for a two-tailed t-test at the specified significance level (α = 0.01) and degrees of freedom. The degrees of freedom can be approximated using the formula:

df = (sA^2 / nA + sB^2 / nB)^2 / ((sA^2 / nA)^2 / (nA - 1) + (sB^2 / nB)^2 / (nB - 1))

Using the given values, we can calculate the degrees of freedom (df).

Next, we can find the critical value for a two-tailed test with the calculated degrees of freedom and the desired significance level (α = 0.01). This can be obtained from a t-table or statistical software.

Step 5: Make a decision:
