## Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Ans= Matplotlib is a popular Python library used for creating static, animated, and interactive visualizations in Python. It provides a wide variety of plots and charts for data visualization purposes. Matplotlib is widely used in various fields such as data science, machine learning, scientific computing, and more.

Here are five plots that can be plotted using the Pyplot module of Matplotlib:

Line plot: Used to visualize data points connected by straight line segments, often used to represent trends over time or other continuous variables.

Scatter plot: Used to display the relationship between two variables by plotting points on a two-dimensional plane.

Histogram: Used to represent the distribution of a single variable by dividing the data into bins and displaying the frequency of occurrences in each bin as bars.

Bar plot: Used to compare the values of different categories or groups by representing them as bars of varying heights.

Pie chart: Used to represent the proportions of different categories within a dataset by dividing a circle into sectors proportional to the values they represent.

## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
import matplotlib.pyplot as plt

plt.xlabel('x axis')

plt.ylabel('y axis')

plt.title('this is title')

plt.scatter(x,y)

plt.show()

Ans= A scatter plot is a type of plot that displays the relationship between two variables by representing individual data points as dots on a two-dimensional plane. It is useful for visualizing the distribution and correlation between two continuous variables.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data for x and y
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Plot scatter plot
plt.scatter(x, y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter Plot Example')
plt.show()
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.Use the following data:
import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

Ans= The subplot() function in Matplotlib is used to create multiple plots in the same figure. It allows you to organize plots in a grid layout, specifying the number of rows and columns, and the position of each plot within the grid. This function is useful when you want to compare different datasets or visualize multiple aspects of the same dataset in one figure.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title('Line 1')

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line 2')

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line 3')

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title('Line 4')

plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

Ans=A bar plot is a type of plot that represents categorical data with rectangular bars. The length of each bar corresponds to the value of the data it represents. Bar plots are commonly used to compare the values of different categories or groups within a dataset.

Bar plots are used for the following purposes:

Comparison: They allow easy comparison of the magnitudes of different categories.

Trend Visualization: They can show trends over time or across different categories.

Distribution: They can display the distribution of data across categories.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='skyblue')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Profit by Company (Bar Plot)')
plt.show()

# Horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='lightgreen')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.title('Profit by Company (Horizontal Bar Plot)')
plt.show()
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

Ans=A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset based on five summary statistics: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. It provides a visual summary of the central tendency, spread, and skewness of the dataset, as well as any potential outliers.

Box plots are used for the following purposes:

Identifying outliers: Box plots help identify potential outliers in the dataset by displaying data points that fall outside the whiskers.

Comparing distributions: They allow easy comparison of the distributions of multiple datasets.

Visualizing variability: They provide insights into the spread and variability of the data.

Detecting skewness: Box plots can reveal the skewness of the data based on the relative positions of the median and quartiles.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create box plot
plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot of Box 1 and Box 2')
plt.grid(True)
plt.show()
```
