## Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans= Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. It offers a wide range of plots for visualizing different types of data. Here are five plots that can be created using Seaborn along with their uses:

Scatter plot:

Uses: Scatter plots are used to visualize the relationship between two continuous variables. They help identify patterns, correlations, and outliers in the data.

Histogram:

Uses: Histograms are used to represent the distribution of a single variable. They provide insights into the central tendency, spread, and skewness of the data.

Bar plot:

Uses: Bar plots are used to compare the values of different categories or groups. They are effective in visualizing categorical data and identifying variations among groups.

Box plot:

Uses: Box plots, also known as box-and-whisker plots, are used to visualize the distribution of a dataset and detect outliers. They provide a summary of the central tendency, spread, and skewness of the data.

Heatmap:

Uses: Heatmaps are used to visualize the correlation between multiple variables in a dataset. They are particularly useful in identifying patterns and relationships in large datasets.

## Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Plot line plot for different events and regions
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region", markers=True)
plt.title("Signal vs Timepoint for Different Events and Regions")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.legend(title="Event")
plt.show()
```

## Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Plot box plots for 'age' and 'fare' against 'pclass'
plt.figure(figsize=(12, 6))

# Box plot for 'age'
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic_data, x='pclass', y='age')
plt.title('Box Plot of Age by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Age')

# Box plot for 'fare'
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic_data, x='pclass', y='fare')
plt.title('Box Plot of Fare by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Fare')

plt.tight_layout()
plt.show()
```

## Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Plot histogram for 'price' with 'cut' as hue
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', kde=True, palette='viridis', multiple='stack')
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend(title='Cut')
plt.show()
```

## Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris_data = sns.load_dataset("iris")

# Plot pair plot with hue for 'species'
sns.pairplot(iris_data, hue='species', palette='viridis')
plt.suptitle('Pair Plot of Iris Dataset with Species Hue')
plt.show()
```

## Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix suitable for heatmap
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, cmap='YlGnBu', annot=True, fmt='d', linewidths=0.5)
plt.title('Passenger Counts by Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
