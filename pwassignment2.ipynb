## Q4. What is a generator function in python? Why yield keyword is used? Give an example of a generator function.

Ans= Generator function return iterator object with sequence of values.Yield keyword control the flow of generator function which is similar to return statement used for returning values in python.
Example:

def fib(n):
    
    a,b=0,1
    
    for i in range(n):
        
        yield a
        
        a,b=b,a+b

for i in fib(10):
